```python
## Jupyter notebook file to compare poisson ratio  (and intercept of fitting) of simulated network for different epsilon values
## This file uses the poisson ratios calculated using test_data_avg_mean_width_tensile.ipynb

## Devosmita Sen##
## August 2024 ##

import numpy as np
from matplotlib import pyplot as plt
eps_arr=[35,40,45,46,50]
cnt=0
for eps in eps_arr:
    ##cnt=cnt+2
    folder='./tol_0.1_eps_'+str(eps)+'_4factor/'
    data=np.genfromtxt(folder+'poisson_ratio_data_y.txt')
    plt.figure(cnt)
    #plt.title('tol=0.01, 4factor')
    plt.axhline(y = 0.5, color = 'k', linestyle = '--') 
    plt.errorbar(data[:,0],data[:,1],yerr=data[:,2],label=str(eps),capsize=5, fmt='o-')
    plt.xlabel('perc_width')
    plt.ylabel('poisson')
    plt.legend()
    data=np.genfromtxt(folder+'b_data_y.txt')
    plt.figure(cnt+1)
    #plt.title('tol=0.01, 4factor')
    plt.axhline(y = 0.0, color = 'k', linestyle = '--') 
    plt.errorbar(data[:,0],data[:,1],yerr=data[:,2],label=str(eps),capsize=5, fmt='o-')
    plt.legend()
    plt.xlabel('perc_width')
    plt.ylabel('b')
    
    
    
    
    
```
